```python
X, y =  make_classification(n_samples=300, n_features=2, n_redundant=0, n_clusters_per_class=1, class_sep=5.0)
```

```python
scatter(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X[:200], y[:200])
y_pred = clf.predict(X[200:]) #otlozhennaya vyborka i predskazanie dlia nee
```

```python
print(y_pred == y[200:])
#true esli kazhduy otvet y_pred sovpadaet so svoim istinnym otvetom
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y[200:], y_pred)
```

```python
#crossvalid
cv = GridSearchCV(LogisticRegression(), param_grid={'c':{1, 10, 100}})
cv.fit(X, y)
cv.grid_scores_
#scores vyvodit kachestvo dlia kazhdogo hyperparametra
#c - coef regularizacii
```
