```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = pd.read_csv('https://raw.githubusercontent.com/iad34/seminars/master/materials/coord.csv')
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(points['LONG'], points['LAT'], alpha=0.5)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
class KMeans:
    def __init__(self, n_clusters, max_iter=100, show=False):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.show = show
    
    def fit_transform(self, X):
        n_samples = X.shape[0]
        n_features = X.shape[1]
        clusters = np.random.randint(0, self.n_clusters, n_samples)
        self.centres = np.zeros((self.n_clusters, n_features))
        
        for it in range(self.max_iter):
            for c in range(self.n_clusters):
                self.centres[c] = X[clusters == c].mean(axis=0)
                #print(self.centres)
                
            distances = pairwise_distances(X, self.centres)
            clusters = np.argmin(distances, axis=1)
            if self.show:
                plt.figure(figsize=(16, 8))
                plt.scatter(X[:, 1], X[:, 0], alpha=0.5, c=clusters)
                plt.show()
    
        return clusters
```

```python
k_means = KMeans(3)
clusters = k_means.fit_transform(points.values)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(points['LONG'], points['LAT'], alpha=0.5, c=clusters)
```

```python
k_means = KMeans(5, max_iter=5, show=True)
clusters = k_means.fit_transform(points.values)
```
